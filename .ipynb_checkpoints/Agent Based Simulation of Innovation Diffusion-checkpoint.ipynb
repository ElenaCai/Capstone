### 1 - Von Neumann SI Model <a class="anchor" id="chapter1"></a>
The Von Neumann SI Model is the starting point of the modeling process. The basic SI Model draws an analogy to the epidemic simulation within the field. It assumes the population has a fixed size of n*n. Each agent can only be in two states for each agent: susceptible (S) to infection or infected/adopted to the idea (I). The primary parameter is p in the basic SI Model. For each update step, two agents are chosen at random, and the infection takes place at a probability of p. This section uses the Von-Neumann neighborhood to the SI Model to capture the neighborhood effect. Communication only takes place between neighbors. The goal is to consider real-life social connections and better visualization of the hub effect. 

```python
class vonNeumannSIModel():
    
    def __init__(self, n, p):
        self.n = n #size of the space n*n
        self.p = p #infectibility parameter
        self.count = 0 #number of adopted agents
        self.window = np.zeros([self.n, self.n]) #create grid
    
    #initialize
    def initialize(self):
        #build a window of zie n*n
        self.window = np.zeros([self.n, self.n])
        
    #visualize the system
    def observe(self):
        plt.cla()
        plt.axis('off')
        plt.imshow(self.window)
        
    #choose neighbor for the agent at [x,y]
    def select_neighbor(self,x,y):
        r = random.randint(0,3) #randomly choose a side
        
        #select top neighbor
        if r == 0:
            return x, (y-1)%self.n
        
        #select bottom neighbor
        if r == 1:
            return x, (y+1)%self.n 
        
        #select left neighbor
        if r == 2:
            return (x-1)%self.n, y
        
        #select right neighbor
        if r == 3:
            return (x+1)%self.n, y
        
    #randomly select an agents and communicate to its neighbor 
    def update(self):
        agent1 = [random.randint(0,self.n-1),random.randint(0,self.n-1)] 
        agent2 = self.select_neighbor(agent1[0],agent1[1]) 
        
        #if one of the agent is not yet adopted,
        #adopt with % of the infectibility parameter
        if self.window[agent1[0],agent1[1]] != self.window[agent2[0],agent2[1]]: 
        
            r = random.random()
            if r < self.p:

                self.window[agent1[0],agent1[1]] = 1
                self.window[agent2[0],agent2[1]] = 1
                self.count += 1
```

```python
def visualizeVonNeumannSIModel(p):
    #set up a 10*10 population
    #with p=100%
    vonNeumannSIModel1 = vonNeumannSIModel(n = 10, p = p)

    #count how many adoptors there are
    count1 = []

    #make one agent adopted to an innovation
    vonNeumannSIModel1.window[4,4] = 1
    vonNeumannSIModel1.count += 1
    count1.append(vonNeumannSIModel1.count)

    #Build a 9*9 grids to present the output
    fig, axs = plt.subplots(3, 3, dpi=200, figsize=(15,15))
    axs[0,0].imshow(vonNeumannSIModel1.window)
    axs[0,0].axis('off')
    axs[0,0].set_title("Step: 0")

    for i in range(1,9):
        for j in range(50):
            vonNeumannSIModel1.update()
            count1.append(vonNeumannSIModel1.count)

        axs[int(i/3), i%3].imshow(vonNeumannSIModel1.window)
        axs[int(i/3), i%3].axis('off')
        axs[int(i/3), i%3].set_title('Step: ' + str(i*50))

    print("Number of adopted agent =",vonNeumannSIModel1.count)
    print("Percentage of adopted agent =",vonNeumannSIModel1.count/100)
    
    plt.figure(dpi=100)
    plt.ylim(0, 100)
    plt.xlabel("time steps")
    plt.ylabel("number of adoptors")
    plt.title("p = " + str(p))
    plt.plot(count1)
```

### 2 - Updatable Multi-Innovation Model  <a class="anchor" id="chapter2"></a>
This model introduces the main parameter p_u, the probability of an idea update. The goal of introducing p_u is to capture the phenomenon of innovation evolution and innovation competition. 

At each update, an idea update happens with the probability of p_u. A random agent is chosen during an idea update, and its current adopted idea is updated to a new idea. For example, if there were 10 ideas in the system, and the agent currently adopts the idea with index 3, then the agent will adopt the idea with index 11 instead of 3 after the update, and the new idea with index 11 will be stored. If the chosen agent for idea update is not yet adopting any idea, it generates a new idea. This is necessary to avoid empty updates. 

The limitation of this method is that there is no recording of update connections. Instead of updating from 3.1 to 3.2, each idea is updated to an unrelated index. The advantage of this method is that it helps with clear visualization of changes. In addition, it is simple for indexing. Also, it captures a phenomenon when an idea is updated to the extent that it seems unrelated to its origin.

For 1 – pu of the time, an idea transmission will take place. This process is the same comparing to the previous Von Neumann SI Model. The infectibility parameter is set at 100% to all the following simulations to study the effect of the p_u parameter. 

```python
class UIMWIG():
    
    def __init__(self, n, pu):
        self.n = n #size of the space n*n
        self.pu = pu #update probability parameter
        self.count = []
        self.window = np.zeros([self.n, self.n]) #create grid
    
    #initialize
    def initialize(self):
        #build a window of zie n*n
        self.window = np.zeros([self.n, self.n])
        
    #visualize the system
    def observe(self):
        plt.cla()
        plt.axis('off')
        plt.imshow(self.window)
        
    #choose neighbor for the agent at [x,y]
    def select_neighbor(self,x,y):
        r = random.randint(0,3) #randomly choose a side
        
        #select top neighbor
        if r == 0:
            return x, (y-1)%self.n
        
        #select bottom neighbor
        if r == 1:
            return x, (y+1)%self.n 
        
        #select left neighbor
        if r == 2:
            return (x-1)%self.n, y
        
        #select right neighbor
        if r == 3:
            return (x+1)%self.n, y
        
    def innovationUpdate(self):
        #randomly choose an agent to update its currently adopted innovation
        agent = [random.randint(0,self.n-1),random.randint(0,self.n-1)] 

        #check whether the chosen agent has adopted to any innovation
        if self.window[agent[0],agent[1]] != 0:
            #update innovation
            self.count[int(self.window[agent[0],agent[1]])-1]-=1
            self.window[agent[0],agent[1]] = len(self.count)+1
            self.count.append(1)
        
        #if the chosen agent has never adopted to any innovation, generate one
        if self.window[agent[0],agent[1]] == 0:
            self.window[agent[0],agent[1]] = len(self.count)+1
            self.count.append(1)


    def innovationCommunication(self):
        #randomly choose an agent as communicator
        agent1 = [random.randint(0,self.n-1),random.randint(0,self.n-1)] 
        
        #check whether it already has adopted to an innovation
        if int(self.window[agent1[0],agent1[1]]) != 0:
        
            #randomly select a neighbor
            agent2 = self.select_neighbor(agent1[0],agent1[1]) 

            #check which innovation does agent 1 adopt right now
            innovationIndex = int(self.window[agent1[0],agent1[1]])

            #update innovation count
            self.count[int(self.window[agent1[0],agent1[1]])-1]+=1
            if int(self.window[agent2[0],agent2[1]]) != 0:
                self.count[int(self.window[agent2[0],agent2[1]])-1]-=1
            self.window[agent2[0],agent2[1]] = innovationIndex

        #check whether it already has adopted to an innovation
        if int(self.window[agent1[0],agent1[1]]) == 0:
            self.innovationCommunication()
    
    #innovation update 
    def update(self):
        
        #for p_u%, innovation update happens
        p = random.random()
        if p < self.pu:
            self.innovationUpdate()

        #for 1-p_u%, communication happens between neighbors
        else:
            self.innovationCommunication()
```

### 3 - Updatable Innovations Model with Normal Infectibility Parameter Distribution   <a class="anchor" id="chapter3"></a>
This model iterates on the previous Updatable Muti-Innovation Model by adding distribution to the infectibility parameter. In the last section, the infectibility parameter was kept at 100% to study the effect of p_u. This section aims to incorporate randomness on the probability of adoption based on Rogers Theory of Innovation Diffusion. The theory indicates that people’s willingness to accept new ideas follows an approximately normal distribution, previously introduced in Figure 2.

### 3.1 - Distribution for Infectibility Parameter   <a class="anchor" id="chapter3.1"></a>
Different from an actual normal distribution, the adoption catagorization distribution has a limit from 0% to 100%. The following cells aim to mimic the distribution given by Roger's Thoery of Innovation Diffusion (last graph).

```python
plst = []
for x in range(1000000):
    p = np.random.normal(loc=0.5, scale=0.2)
    #for any p value within the limit, make no change
    if p>=0 and p<=1:
        plst.append(p)
    #for any negative value, flip it to positive
    if p<0 and (-p)<=0.1:
        plst.append(-p)
    #for any value greater than 100%, flip it over the 100%
    if p>1 and p-1<=0.1:
        plst.append(2-p)
        
plt.figure(dpi=100)
plt.xlabel("p")
plt.ylabel("Frequency")
plt.title("Probability Distribution of Infectibility Parameter")
plt.hist(plst,bins='auto')
```

The following cell prints out the number of adoptor to each catagory. The catagoriztaion is accorate to the first digit of all catagories.

```python
innovatorCount = 0
earlyAdopterCount = 0
earlyMajorityCount = 0
lateMajorityCount = 0
laggardsCount = 0

for i in plst:
    if i<0.1:
        innovatorCount +=1
    if i>=0.1 and i<0.3:
        earlyAdopterCount +=1
    if i>=0.3 and i<0.5:
        earlyMajorityCount +=1
    if i>=0.5 and i<0.7:
        lateMajorityCount +=1
    if i>=0.7:
        laggardsCount +=1
        
print("Innovator:",innovatorCount/10000,"%")
print("Early Adopter:",earlyAdopterCount/10000,"%")
print("Early Majority:",earlyMajorityCount/10000,"%")
print("Late Majority:",lateMajorityCount/10000,"%")
print("Laggards:",laggardsCount/10000,"%")
```

### 3.2 - Model Implementation   <a class="anchor" id="chapter3.2"></a>
Here, I implement the model incorporating the probability distribution of the infectibility parameter. 

```python
class modelSix():
    
    def __init__(self, n, pu):
        self.n = n #size of the space n*n
        self.pu = pu #update probability parameter
        self.count = []
        self.window = np.zeros([self.n, self.n]) #create grid
    
    #initialize
    def initialize(self):
        #build a window of zie n*n
        self.window = np.zeros([self.n, self.n])
        
    #visualize the system
    def observe(self):
        plt.cla()
        plt.axis('off')
        plt.imshow(self.window)
        
    #choose neighbor for the agent at [x,y]
    def select_neighbor(self,x,y):
        r = random.randint(0,3) #randomly choose a side
        
        #select top neighbor
        if r == 0:
            return x, (y-1)%self.n
        #select bottom neighbor
        if r == 1:
            return x, (y+1)%self.n 
        #select left neighbor
        if r == 2:
            return (x-1)%self.n, y
        #select right neighbor
        if r == 3:
            return (x+1)%self.n, y
        
    def adoptionProbability(self):
        p = np.random.normal(loc=0.5, scale=0.2)
        #for any p value within the limit, make no change
        if p>=0 and p<=1:
            return(p)
        #for any negative value, flip it to positive
        if p<0 and (-p)<=0.1:
            return(-p)
        #for any value greater than 100%, flip it over the 100%
        if p>1 and p-1<=0.1:
            return(2-p)
        else:
            return(self.adoptionProbability())
    
    def innovationUpdate(self):
        #randomly choose an agent to update its currently adopted innovation
        agent = [random.randint(0,self.n-1),random.randint(0,self.n-1)] 

        #check whether the chosen agent has adopted to any innovation
        if self.window[agent[0],agent[1]] != 0:
            #update innovation
            self.count[int(self.window[agent[0],agent[1]])-1]-=1
            self.window[agent[0],agent[1]] = len(self.count)+1
            self.count.append(1)
        
        #if the chosen agent has never adopted to any innovation, generate one
        if self.window[agent[0],agent[1]] == 0:
            self.window[agent[0],agent[1]] = len(self.count)+1
            self.count.append(1)


    def innovationCommunication(self):
        #randomly choose an agent as communicator
        agent1 = [random.randint(0,self.n-1),random.randint(0,self.n-1)] 
        
        #check whether it already has adopted to an innovation
        if int(self.window[agent1[0],agent1[1]]) == 0:
            self.innovationCommunication()
        
        #adoption happens with a probability following the adoption probability distribution
        r = random.random()
        p = self.adoptionProbability()
        if r<p:

            #check whether it already has adopted to an innovation
            if int(self.window[agent1[0],agent1[1]]) != 0:

                #randomly select a neighbor
                agent2 = self.select_neighbor(agent1[0],agent1[1]) 

                #check which innovation does agent 1 adopt right now
                innovationIndex = int(self.window[agent1[0],agent1[1]])

                #update innovation count
                self.count[int(self.window[agent1[0],agent1[1]])-1]+=1
                if int(self.window[agent2[0],agent2[1]]) != 0:
                    self.count[int(self.window[agent2[0],agent2[1]])-1]-=1
                self.window[agent2[0],agent2[1]] = innovationIndex


    #innovation update 
    def update(self):
        
        #for p_u%, innovation update happens
        p = random.random()
        if p < self.pu:
            self.innovationUpdate()

        #for 1-p_u%, communication happens between neighbors
        else:
            self.innovationCommunication()
```

### 4 - Innovation Impact Analysis <a class="anchor" id="chapter4"></a> 

Until now, the "real innovation" is not yet defined through code. An idea is assumed to pre-exist in the system, and any agent can update the idea into a new version. However, not any idea/creation is an innovation. Most ideas die out before their implementation. Similarly, most products do not survive through the entire product life cycle. This section aims to study the diffusion of innovation, taking into account the importance of impact - a unique new idea is only publically referred to as an innovation if it has a significant impact.

One way to define innovation impact is to use a certain threshold of percentage adoption p_i. If the idea is influencing over p_i*n*n number of agents within a population at a time step, it is categorized as a living innovation. 

The "count" parameter of the model counts the number of ideas present in the system. Over time, some ideas are born, but some are dead. To capture the system's dynamic, we need to capture how long each idea lives and how much impact it has made. Hence, I create another "innovationCount" parameter to store the information. 

```python
class impactAnalysis():
    
    def __init__(self, n, pu, pi):
        self.n = n #size of the space n*n
        self.pu = pu #update probability parameter
        self.pi = pi #innovation catagorization threshold
        self.count = []
        self.innovationCount = {} #count number of innovation at each time step
        self.window = np.zeros([self.n, self.n]) #create grid
    
    #initialize
    def initialize(self):
        #build a window of zie n*n
        self.window = np.zeros([self.n, self.n])
        
    #visualize the system
    def observe(self):
        plt.cla()
        plt.axis('off')
        plt.imshow(self.window)
        
    #choose neighbor for the agent at [x,y]
    def select_neighbor(self,x,y):
        r = random.randint(0,3) #randomly choose a side
        
        #select top neighbor
        if r == 0:
            return x, (y-1)%self.n
        #select bottom neighbor
        if r == 1:
            return x, (y+1)%self.n 
        #select left neighbor
        if r == 2:
            return (x-1)%self.n, y
        #select right neighbor
        if r == 3:
            return (x+1)%self.n, y
        
    def adoptionProbability(self):
        p = np.random.normal(loc=0.5, scale=0.2)
        #for any p value within the limit, make no change
        if p>=0 and p<=1:
            return(p)
        #for any negative value, flip it to positive
        if p<0 and (-p)<=0.1:
            return(-p)
        #for any value greater than 100%, flip it over the 100%
        if p>1 and p-1<=0.1:
            return(2-p)
        else:
            return(self.adoptionProbability())
    
    def innovationUpdate(self):
        #randomly choose an agent to update its currently adopted innovation
        agent = [random.randint(0,self.n-1),random.randint(0,self.n-1)] 

        #check whether the chosen agent has adopted to any innovation
        if self.window[agent[0],agent[1]] != 0:
            #update innovation
            self.count[int(self.window[agent[0],agent[1]])-1]-=1
            self.window[agent[0],agent[1]] = len(self.count)+1
            self.count.append(1)
        
        #if the chosen agent has never adopted to any innovation, generate one
        if self.window[agent[0],agent[1]] == 0:
            self.window[agent[0],agent[1]] = len(self.count)+1
            self.count.append(1)


    def innovationCommunication(self):
        #randomly choose an agent as communicator
        agent1 = [random.randint(0,self.n-1),random.randint(0,self.n-1)] 
        
        #check whether it already has adopted to an innovation
        if int(self.window[agent1[0],agent1[1]]) == 0:
            self.innovationCommunication()
        
        #adoption happens with a probability following the adoption probability distribution
        r = random.random()
        p = self.adoptionProbability()
        if r<p:
        
            #check whether it already has adopted to an innovation
            if int(self.window[agent1[0],agent1[1]]) != 0:

                #randomly select a neighbor
                agent2 = self.select_neighbor(agent1[0],agent1[1]) 

                #check which innovation does agent 1 adopt right now
                innovationIndex = int(self.window[agent1[0],agent1[1]])

                #update innovation count
                self.count[int(self.window[agent1[0],agent1[1]])-1]+=1
                if int(self.window[agent2[0],agent2[1]]) != 0:
                    self.count[int(self.window[agent2[0],agent2[1]])-1]-=1
                self.window[agent2[0],agent2[1]] = innovationIndex

    #innovation update 
    def update(self):
        
        #for p_u%, innovation update happens
        p = random.random()
        if p < self.pu:
            self.innovationUpdate()

        #for 1-p_u%, communication happens between neighbors
        else:
            self.innovationCommunication()
            
        #catagorize innovation
        for i in range(len(self.count)):
            if self.count[i]>=self.pi*self.n*self.n:
                if i+1 in self.innovationCount:
                    self.innovationCount[i+1]+=1
                else:
                    self.innovationCount[i+1]=1      
```

### 5 - Single Agent Alternization <a class="anchor" id="chapter5"></a> 
This section studies the extent to which altering a single agent can cause an overall impact on the population. This new model alters the top left agent of the grid with additional parameters of pu_x and pc_x. These parameters represent how much more frequently the special agent performs an innovation update/communication comparing the general population. For example, x0 means the agent performs as frequently as the general population. x1 means the agent performs one time more frequently than the general population. 

```python
class singleAgentAlternation():
    
    def __init__(self, n, pu, pi, pu_x, pc_x):
                
        self.n = n #size of the space n*n
        self.pu = pu #update probability parameter
        self.pi = pi #innovation catagorization threshold
        self.pu_x = pu_x/(n*n) #how much more frequent does the special agent perform an innovation update
        self.pc_x = pc_x/(n*n) #how much more frequent does the special agent perform an innovation communication
        
        #pu_x and pc_x ≥ 0. When they equal to 0, the special agent performs the same as other agent. 
        #When they equal 1, they perform 1 time more likely than ordinary agent (1X more)

        self.count = []
        self.innovationCount = {}
        self.window = np.zeros([self.n, self.n]) #create grid
    
    #initialize
    def initialize(self):
        #build a window of zie n*n
        self.window = np.zeros([self.n, self.n])
        
    #visualize the system
    def observe(self):
        plt.cla()
        plt.axis('off')
        plt.imshow(self.window)
        
    #choose neighbor for the agent at [x,y]
    def select_neighbor(self,x,y):
        r = random.randint(0,3) #randomly choose a side
        
        #select top neighbor
        if r == 0:
            return x, (y-1)%self.n
        #select bottom neighbor
        if r == 1:
            return x, (y+1)%self.n 
        #select left neighbor
        if r == 2:
            return (x-1)%self.n, y
        #select right neighbor
        if r == 3:
            return (x+1)%self.n, y
        
    def adoptionProbability(self):
        p = np.random.normal(loc=0.5, scale=0.2)
        #for any p value within the limit, make no change
        if p>=0 and p<=1:
            return(p)
        #for any negative value, flip it to positive
        if p<0 and (-p)<=0.1:
            return(-p)
        #for any value greater than 100%, flip it over the 100%
        if p>1 and p-1<=0.1:
            return(2-p)
        else:
            return(self.adoptionProbability())
    
    def innovationUpdate(self):
        
        #randomly choose an agent to update its currently adopted innovation
        agent = [random.randint(0,self.n-1),random.randint(0,self.n-1)] 

        #check whether the chosen agent has adopted to any innovation
        if self.window[agent[0],agent[1]] != 0:
            #update innovation
            self.count[int(self.window[agent[0],agent[1]])-1]-=1
            self.window[agent[0],agent[1]] = len(self.count)+1
            self.count.append(1)
        
        #if the chosen agent has never adopted to any innovation, generate one
        if self.window[agent[0],agent[1]] == 0:
            self.window[agent[0],agent[1]] = len(self.count)+1
            self.count.append(1)
          
        #update the special agent
        for i in range(np.math.floor(self.pu_x)):
            agent_s = [0,0] #set up the special agent
            
            #check whether the special agent has adopted to any innovation
            if self.window[agent_s[0],agent_s[1]] != 0:
                #update innovation
                self.count[int(self.window[agent_s[0],agent_s[1]])-1]-=1
                self.window[agent_s[0],agent_s[1]] = len(self.count)+1
                self.count.append(1)

            #if the chosen agent has never adopted to any innovation, generate one
            if self.window[agent_s[0],agent_s[1]] == 0:
                self.window[agent_s[0],agent_s[1]] = len(self.count)+1
                self.count.append(1)
                                    
        #update the special agent
        pu = self.pu_x - np.math.floor(self.pu_x)
        ru = random.random()
        if ru < pu:
            agent_s = [0,0] #set up the special agent
            
            #check whether the special agent has adopted to any innovation
            if self.window[agent_s[0],agent_s[1]] != 0:
                #update innovation
                self.count[int(self.window[agent_s[0],agent_s[1]])-1]-=1
                self.window[agent_s[0],agent_s[1]] = len(self.count)+1
                self.count.append(1)

            #if the chosen agent has never adopted to any innovation, generate one
            if self.window[agent_s[0],agent_s[1]] == 0:
                self.window[agent_s[0],agent_s[1]] = len(self.count)+1
                self.count.append(1)  
                            
    def innovationCommunication(self):
        
        #randomly choose an agent as communicator
        agent1 = [random.randint(0,self.n-1),random.randint(0,self.n-1)] 
        
        #check whether it already has adopted to an innovation
        #only communicate when it has adopted to an innovation
        if int(self.window[agent1[0],agent1[1]]) != 0:

        
            #adoption happens with a probability following the adoption probability distribution
            r = random.random()
            p = self.adoptionProbability()
            if r<p:

                #randomly select a neighbor
                agent2 = self.select_neighbor(agent1[0],agent1[1]) 

                #check which innovation does agent 1 adopt right now
                innovationIndex = int(self.window[agent1[0],agent1[1]])

                #update innovation count
                self.count[int(self.window[agent1[0],agent1[1]])-1]+=1
                if int(self.window[agent2[0],agent2[1]]) != 0:
                    self.count[int(self.window[agent2[0],agent2[1]])-1]-=1
                self.window[agent2[0],agent2[1]] = innovationIndex
                
        #update the special agent
        for i in range(np.math.floor(self.pc_x)):
            agent_s = [0,0] #set up the special agent
            
            #check whether the special agent has adopted to any innovation
            #only update when it has adopted to an innovation
            if self.window[agent_s[0],agent_s[1]] != 0:
                #randomly select a neighbor
                agent2 = self.select_neighbor(agent_s[0],agent_s[1]) 

                #adoption happens with a probability following the adoption probability distribution
                r = random.random()
                p = self.adoptionProbability()
                
                if r<p:
                    #randomly select a neighbor
                    agent2 = self.select_neighbor(agent_s[0],agent_s[1]) 

                    #check which innovation does agent_s adopt right now
                    innovationIndex = int(self.window[agent_s[0],agent_s[1]])

                    #update innovation count
                    self.count[int(self.window[agent_s[0],agent_s[1]])-1]+=1
                    if int(self.window[agent2[0],agent2[1]]) != 0:
                        self.count[int(self.window[agent2[0],agent2[1]])-1]-=1
                    self.window[agent2[0],agent2[1]] = innovationIndex
                                    
        #update the special agent
        pc = self.pc_x - np.math.floor(self.pc_x)
        rc = random.random()
        if rc < pc:
        
            agent_s = [0,0] #set up the special agent
            
            #check whether the special agent has adopted to any innovation
            if self.window[agent_s[0],agent_s[1]] != 0:
                #randomly select a neighbor
                agent2 = self.select_neighbor(agent1[0],agent1[1]) 

                #check which innovation does agent 1 adopt right now
                innovationIndex = int(self.window[agent1[0],agent1[1]])

                #update innovation count
                self.count[int(self.window[agent1[0],agent1[1]])-1]+=1
                if int(self.window[agent2[0],agent2[1]]) != 0:
                    self.count[int(self.window[agent2[0],agent2[1]])-1]-=1
                self.window[agent2[0],agent2[1]] = innovationIndex 
                                
    #innovation update 
    def update(self):
        
        #for p_u%, innovation update happens
        p = random.random()
        if p < self.pu:
            self.innovationUpdate()

        #for 1-p_u%, communication happens between neighbors
        else:
            self.innovationCommunication()
            
        #catagorize innovation
        for i in range(len(self.count)):
            if self.count[i]>=self.pi*self.n*self.n:
                if i+1 in self.innovationCount:
                    self.innovationCount[i+1]+=1
                else:
                    self.innovationCount[i+1]=1      
```

### 5.1 - Setting pu=10%, pi=5%, pu_x=0x, pc_x=0x <a class="anchor" id="chapter5.1"></a> 
This section sets the update rate of the system to 10%, and innovation definition bar at 5%. 
The special agent is set to the same as originary agents with 0x more update or more communication.

```python
def visualizeSingleAgentAlternation(pu, pi, pu_x, pc_x):
    #set up a 10*10 population
    singleAgentAlternation1 = singleAgentAlternation(n=10, pu=pu, pi=pi, pu_x=pu_x, pc_x=pc_x)
   
    #create an innovation
    singleAgentAlternation1.window[2,2] = 1
    singleAgentAlternation1.count.append(1)

    maxAdoptionLst = [1]
    maxInnovationIndexLst = [1]

    #Build a 9*9 grids to present the output
    fig, axs = plt.subplots(3, 3, dpi=200, figsize=(15,15))
    axs[0,0].imshow(singleAgentAlternation1.window)
    axs[0,0].axis('off')
    axs[0,0].set_title("Step: 0")

    for i in range(1,9):
        for j in range(300):
            singleAgentAlternation1.update()
            maxAdoption = max(singleAgentAlternation1.count)
            maxAdoptionLst.append(maxAdoption)
            maxInnovationIndexLst.append(singleAgentAlternation1.count.index(maxAdoption)+1)

        axs[int(i/3), i%3].imshow(singleAgentAlternation1.window)
        axs[int(i/3), i%3].axis('off')
        axs[int(i/3), i%3].set_title('Step: ' + str(i*300))

    print(singleAgentAlternation1.window)
    print("At the last time step, the innovation with index", maxInnovationIndexLst[-1], "has the maximum adoption number of", maxAdoptionLst[-1])
    print("Througout the time,", len(singleAgentAlternation1.innovationCount), "innovations emerge from the system")

    return(len(singleAgentAlternation1.innovationCount))
```

### 5.5 - Influence of Special Agent Update Parameter <a class="anchor" id="chapter5.5"></a> 
This section studies the extent to which can the special agent influence the system by updating more frequently. 

```python
def singleAgentAlternation_i(pu, pi, pu_x, pc_x):
    #set up a 10*10 population
    singleAgentAlternation1 = singleAgentAlternation(n=10, pu=pu, pi=pi, pu_x=pu_x, pc_x=pc_x)
   
    #create an innovation
    singleAgentAlternation1.window[2,2] = 1
    singleAgentAlternation1.count.append(1)

    maxAdoptionLst = [1]
    maxInnovationIndexLst = [1]

    for i in range(1,9):
        for j in range(300):
            singleAgentAlternation1.update()
            maxAdoption = max(singleAgentAlternation1.count)
            maxAdoptionLst.append(maxAdoption)
            maxInnovationIndexLst.append(singleAgentAlternation1.count.index(maxAdoption)+1)

    return(len(singleAgentAlternation1.innovationCount))
```

```python
def impactAnalysisSAA_pu():
    innovationCountLst = []
    
    pu_x = np.linspace(0,100,101)
    for pu_x in pu_x: 
        count = singleAgentAlternation_i(0.1, 0.05, pu_x, 0)
        innovationCountLst.append(count)
    return(innovationCountLst)
```

```python
a = impactAnalysisSAA_pu()
m_1, k_1 = np.polyfit(np.linspace(0,100,101), a, 1)
```

To study the effect of pu_x on the system dynamics, other parameters are set to be p_u = 10%, p_i = 5%, pc_i = 0x. pu_x values ranging from 0x to 100x are used. The following graph plots the number of innovations evolved from the system when the special agents have different update frequency comparing to the general population. The graph does not indicate a strong correlation. It shows a week positive correlation between pu_x and the number of innovations. The more frequently the agent update ideas, the slightly more innovation the system is likely to have. 

```python
plt.figure(dpi=100)
plt.xlabel("pu_x")
plt.ylabel("number of innovation evolved from the system")
plt.title("influence of special agent on the system by altering pu_x")
plt.plot(np.linspace(0,100,101), m_1*np.linspace(0,100,101) + k_1)
plt.scatter(np.linspace(0,100,101),a, s=2)
print("i = ",m_1,"*pu_x +", k_1)
```

### 6 - Flexible Model <a class="anchor" id="chapter6"></a>

```python
#import necessary packages
import matplotlib
import random
import scipy
import pylab as plt
import math
import numpy as np

#builing block for the flexible model
class model(): 
    
    def __init__(self, n, pu, pi):
        self.n = n #size of the space n*n
        self.pu = pu #update probability parameter
        self.pi = pi #innovation catagorization threshold
        self.count = []
        self.innovationCount = {}
        self.window = np.zeros([self.n, self.n]) #create grid   

        #initialize
    def initialize(self):
        #build a window of zie n*n
        self.window = np.zeros([self.n, self.n])
        
    #visualize the system
    def observe(self):
        plt.cla()
        plt.axis('off')
        plt.imshow(self.window)
        
    #choose neighbor for the agent at [x,y]
    def select_neighbor(self,x,y):
        r = random.randint(0,3) #randomly choose a side
        
        #select top neighbor
        if r == 0:
            return x, (y-1)%self.n
        #select bottom neighbor
        if r == 1:
            return x, (y+1)%self.n 
        #select left neighbor
        if r == 2:
            return (x-1)%self.n, y
        #select right neighbor
        if r == 3:
            return (x+1)%self.n, y
        
    def adoptionProbability(self):
        p = np.random.normal(loc=0.5, scale=0.2)
        #for any p value within the limit, make no change
        if p>=0 and p<=1:
            return(p)
        #for any negative value, flip it to positive
        if p<0 and (-p)<=0.1:
            return(-p)
        #for any value greater than 100%, flip it over the 100%
        if p>1 and p-1<=0.1:
            return(2-p)
        else:
            return(self.adoptionProbability())
    
    def innovationUpdate(self):
        #randomly choose an agent to update its currently adopted innovation
        agent = [random.randint(0,self.n-1),random.randint(0,self.n-1)] 

        #check whether the chosen agent has adopted to any innovation
        if self.window[agent[0],agent[1]] != 0:
            #update innovation
            self.count[int(self.window[agent[0],agent[1]])-1]-=1
            self.window[agent[0],agent[1]] = len(self.count)+1
            self.count.append(1)
        
        #if the chosen agent has never adopted to any innovation, generate one
        if self.window[agent[0],agent[1]] == 0:
            self.window[agent[0],agent[1]] = len(self.count)+1
            self.count.append(1)


    def innovationCommunication(self):
        #randomly choose an agent as communicator
        agent1 = [random.randint(0,self.n-1),random.randint(0,self.n-1)] 
        
        #check whether it already has adopted to an innovation
        if int(self.window[agent1[0],agent1[1]]) != 0:
        
            #adoption happens with a probability following the adoption probability distribution
            r = random.random()
            p = self.adoptionProbability()
            if r<p:

                #randomly select a neighbor
                agent2 = self.select_neighbor(agent1[0],agent1[1]) 

                #check which innovation does agent 1 adopt right now
                innovationIndex = int(self.window[agent1[0],agent1[1]])

                #update innovation count
                self.count[int(self.window[agent1[0],agent1[1]])-1]+=1
                if int(self.window[agent2[0],agent2[1]]) != 0:
                    self.count[int(self.window[agent2[0],agent2[1]])-1]-=1
                self.window[agent2[0],agent2[1]] = innovationIndex

    #innovation update 
    def update(self):
        
        #for p_u%, innovation update happens
        p = random.random()
        if p < self.pu:
            self.innovationUpdate()

        #for 1-p_u%, communication happens between neighbors
        else:
            self.innovationCommunication()
            
        #catagorize innovation
        for i in range(len(self.count)):
            if self.count[i]>=self.pi*self.n*self.n:
                if i+1 in self.innovationCount:
                    self.innovationCount[i+1]+=1
                else:
                    self.innovationCount[i+1]=1      
    
def flexibleModel():
    n = int(input("Population Dimension n: "))
    pu = int(input("Update Probability (%): "))/100
    pi = int(input("Innovation Impact Threshold (%): "))/100
    
    #set up initial condition
    model1 = model(n=n, pu=pu, pi=pi)
    model1.window[2,2] = 1
    model1.count.append(1)
    return(model1)

def visualizeYourModel():

    n = int(input("number of steps you'd like to see in each update: "))
    
    #Build a 9*9 grids to present the output
    fig, axs = plt.subplots(3, 3, dpi=200, figsize=(15,15))
    axs[0,0].imshow(yourModel.window)
    axs[0,0].axis('off')
    axs[0,0].set_title("Step: 0")

    for i in range(1,9):
        for j in range(n):
            yourModel.update()

        axs[int(i/3), i%3].imshow(yourModel.window)
        axs[int(i/3), i%3].axis('off')
        axs[int(i/3), i%3].set_title('Step: ' + str(i*n))

    print(yourModel.window)
    print("Througout the time,", len(yourModel.innovationCount), "innovations emerge from the system")
```
